```python
# Partidas

nikoPartidas = pd.read_csv("nikofasedegruposmdl.csv")
gabbiPartidas = pd.read_csv("gabbifasedegruposmdl.csv")
rtzPartidas = pd.read_csv("rtzfasedegruposmdl.csv")
mickePartidas = pd.read_csv("mickefasedegruposmdl.csv")
paparaziPartidas = pd.read_csv("paparazifasedegruposmdl.csv")
scccPartidas = pd.read_csv("scccfasedegruposmdl.csv")
fatedPartidas = pd.read_csv("vtfatedfasedegruposmdl.csv")
dreamPartidas = pd.read_csv("dreamfasedegruposmdl.csv")
iltwPartidas = pd.read_csv("iltwfasedegruposmdl.csv")
savagePartidas = pd.read_csv("23savagefasedegruposmdl.csv")
natsumiPartidas = pd.read_csv("natsumifasedegruposmdl.csv")
envyPartidas = pd.read_csv("envyfasedegruposmdl.csv")
mooPartidas = pd.read_csv("moofasedegruposmdl.csv")
k1Partidas = pd.read_csv("k1fasedegruposmdl.csv")
bernaPartidas = pd.read_csv("bernafasedegruposmdl.csv")
flyflyPartidas = pd.read_csv("flyflyfasedegruposmdl.csv")

# Exemplo das colunas dos data frames

nikoPartidas.head()
```

```python
# Deixar nome e id dos heroes em um dicionario

heroes = pd.read_csv("dotaheroes.csv")
heroes.head()

h = {}

for i in heroes.iterrows():
    h[i[1][0]] = i[1][2]
```

```python
heroesPicks = {}

def heroes(p, matches):
    
    for i in p.iterrows():
        if (i[1][0] in matches):
            if (i[1][6] not in heroesPicks.keys()):
                heroesPicks[i[1][6]] = 1
            else:
                heroesPicks[i[1][6]] += 1
```

```python
heroes(nikoPartidas, mdl)
heroes(gabbiPartidas, mdl)
heroes(rtzPartidas, mdl)
heroes(mickePartidas, mdl)
heroes(paparaziPartidas, mdl)
heroes(scccPartidas, mdl)
heroes(fatedPartidas, mdl)
heroes(dreamPartidas, mdl)
heroes(iltwPartidas, mdl)
heroes(savagePartidas, mdl)
heroes(natsumiPartidas, mdl)
heroes(envyPartidas, mdl)
heroes(mooPartidas, mdl)
heroes(k1Partidas, mdl)
heroes(bernaPartidas, mdl)
heroes(flyflyPartidas, mdl)
```

```python
# Calcula as medias dos principais stats do player

def stats(p, matches): 
    
    nPartidas = 0
    lastHits = 0
    k = 0
    d = 0
    a = 0
    xp = 0
    gp = 0
    
    for i in p.iterrows():
        if (i[1][0] in matches):
            nPartidas += 1
            k += i[1][9]
            d += i[1][10]
            a += i[1][11]
            xp += i[1][13]
            gp += i[1][14]
            lastHits += i[1][18]
            
    kda = ( (k/nPartidas) + (a/nPartidas) ) / (d/nPartidas)
    return [k/nPartidas, d/nPartidas, a/nPartidas, xp/nPartidas, gp/nPartidas, lastHits/nPartidas, kda]
```

```python
# Dicionario com os nomes dos players(keys) e seus stats(values)

playersStats = {}

playersStats['Niko'] = stats(nikoPartidas, mdl)
playersStats['Gabbi'] = stats(gabbiPartidas, mdl)
playersStats['Rtz'] = stats(rtzPartidas, mdl)
playersStats['Micke'] = stats(mickePartidas, mdl)
playersStats['Paparazi'] = stats(paparaziPartidas, mdl)
playersStats['Sccc'] = stats(scccPartidas, mdl)
playersStats['Fated'] = stats(fatedPartidas, mdl)
playersStats['Dream'] = stats(dreamPartidas, mdl)
playersStats['Iltw'] = stats(iltwPartidas, mdl)
playersStats['Savage'] = stats(savagePartidas, mdl)
playersStats['Natsumi'] = stats(natsumiPartidas, mdl)
playersStats['Envy'] = stats(envyPartidas, mdl)
playersStats['Moo'] = stats(mooPartidas, mdl)
playersStats['K1'] = stats(k1Partidas, mdl)
playersStats['Berna'] = stats(bernaPartidas, mdl)
playersStats['FlyFly'] = stats(flyflyPartidas, mdl)
```

```python
# Ordenando do maior KDA para o menor KDA

rank = sorted(playersStats.items(), key=lambda t: t[1][6], reverse=True)

# Alinha a direita com 10 espaÃ§os em branco

print("{0:>10}".format('Jogador'), "{0:>10}".format('KDA'), "{0:>10}".format('K'), "{0:>10}".format('D'),   
      "{0:>10}".format('A'), "{0:>10}".format('XPM'), "{0:>10}".format('OPM'), "{0:>10}".format('LHs'))

# Duas casas decimais

for i in rank:
    print ("{0:>10}".format(i[0]), "{0:>10}".format('%.2f' % i[1][6]), "{0:>10}".format('%.2f' % i[1][0]), 
           "{0:>10}".format('%.2f' % i[1][1]), "{0:>10}".format('%.2f' % i[1][2]), "{0:>10}".format('%.2f' % i[1][3]), 
           "{0:>10}".format('%.2f' % i[1][4]), "{0:>10}".format('%.2f' % i[1][5]))
```
